```python
unique_user = list(sessions.user_id_hash.unique())
```

```python
mapping = list(np.arange(0, 621106))
```

```python
user_id_map = {}
```

```python
for u,m in zip(unique_user, mapping):
    user_id_map[u] = m
```

```python
#user_id_map
```

```python
for i in range(len(new_sessions)):
    hashing_value = new_sessions["user_id_hash"].values[i]
    new_sessions["user_id_hash"].values[i] = user_id_map[hashing_value]
```
